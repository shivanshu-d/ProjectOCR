```python
import sys
from pathlib import Path
import cv2
import pytesseract

sys.path.append("../")

pytesseract.pytesseract.tesseract_cmd = "/opt/homebrew/bin/tesseract"

from src.ocr_engine import OCREngine
from src.preprocessing import Preprocessor

ocr = OCREngine(engine="tesseract")
pre = Preprocessor()

img_path = Path("../datasets/ReverseWayBill/reverseWaybill-160390797970200578_1.jpg")

# Read image
img = cv2.imread(str(img_path))
if img is None:
    raise Exception(f"Could not read image at: {img_path}")

print("Image loaded successfully" )

processed = pre.run_pipeline(img)
print("Preprocessing is done")

results = ocr.recognize(processed)
if not results:
    print("No text detected")
else:
    print("\n===== OCR RESULTS =====\n")
    for r in results:
        print(f"[{r['engine']}] {r['text']} (conf: {r['confidence']})")
```
